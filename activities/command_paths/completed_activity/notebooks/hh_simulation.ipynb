# Demonstration of Hodgkin–Huxley neuron model

This notebook demonstrates the Hodgkin–Huxley (HH) model (1952), which describes
how neurons produce action potentials.

The model is a four dimensional differential equation. The dimensions are the membrane voltage $V$ and three dimensionless gating variables ($m$, $h$, $n$) that govern sodium and potassium channel openings.

A constant applied current of $I_{app} = 10\ \mu\text{A/cm}^2$ drives the cell into a regularly-spiking regime, producing a train of action potentials.

```python
import sys
import os

# Allow importing from the repo root's utils/ package
sys.path.insert(0, os.path.join(os.path.dirname(os.path.abspath('__file__')), '..'))

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

from utils import hh_model

%matplotlib inline
mpl.rcParams.update({'font.size': 12, 'axes.spines.top': False, 'axes.spines.right': False})
```

### Model parameters and simulation settings

```python
params = hh_model.get_default_params() # Defines the cell properties

I_app   = 10.0      # applied current (uA/cm2)
t_start =  0.0      # ms
t_end   = 100.0     # ms
dt      =  0.025    # ms
```

### Run the simulation

```python
sol = hh_model.simulate(
    t_span=(t_start, t_end),
    I_app=I_app,
    params=params,
    dt=dt,
)

t = sol.t          # ms
V = sol.y[0]       # membrane potential (mV)
m = sol.y[1]       # Na activation gate
h = sol.y[2]       # Na inactivation gate
n = sol.y[3]       # K activation gate

print(f'Simulated {t_end - t_start:.0f} ms, {len(t)} time points')
print(f'Peak voltage: {V.max():.1f} mV, Trough: {V.min():.1f} mV')
```

### Membrane potential trace

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(t, V, linewidth=1.2)

ax.set_xlabel('Time (msec)', fontsize=13)
ax.set_ylabel(r'Membrane potential (mV)', fontsize=13)
ax.set_title(f'Hodgkin–Huxley neuron simulation for $I$ = {I_app} $mA/cm^2$', fontsize=14)
ax.set_xlim(t_start, t_end)

plt.tight_layout()
#plt.savefig('hh_trace.pdf', dpi=150, bbox_inches='tight')
plt.show()
```

### Gating variables

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 6), sharex=True)

axes[0].plot(t, V, linewidth=1.2, label='V')
axes[0].set_ylabel('Voltage (mV)', fontsize=12)
axes[0].set_title('Membrane potential and gating variables', fontsize=13)

axes[1].plot(t, m, label='m', linewidth=1.1)
axes[1].plot(t, h, label='h', linewidth=1.1)
axes[1].plot(t, n, label='n', linewidth=1.1)
axes[1].set_ylabel('Gating', fontsize=12)
axes[1].set_xlabel('Time (msec)', fontsize=12)
axes[1].legend(loc='upper right', fontsize=10)
axes[1].set_ylim(-0.05, 1.05)

for ax in axes:
    ax.set_xlim(t_start, t_end)

plt.tight_layout()
plt.show()
```
