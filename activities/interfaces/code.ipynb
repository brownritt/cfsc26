# Activity: The code you see, the code that runs

In this notebook, we exhibited the difference between interfaces (text editors, notebooks) and the python kernels that 

Notebooks naturally use a Read Eval Print Loop (REPL) design. When you execute a cell, it gets sent to the kernel, the kernel does the work, and the output is returned to the notebook.

Scripts generally execute in their own shell, all at once. However, one can turn on "cell mode" is VS Code (and other applications like Matlab, or R if using `Rmd` notebooks) to get a kind of "throw away notebook" to temporarily work with subsections of code.

```python
# If the cells are executed out of order, you might not get what you expect !
# Good practice is to complete a chunk of work, restart the kernel, and
# then run the cells in order to make sure that the visible state and the 
# hidden python kernel state are synchronized
 
print(a)
```
